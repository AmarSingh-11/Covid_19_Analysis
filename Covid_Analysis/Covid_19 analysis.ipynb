```python
# CHECKING FOR THE MISSING VALUES 
missing_values = df.isnull().sum()
# DISPLAYING THE COUNTS OF MISSING VALUES FOR EACH VALUES
missing_values
```

```python
 #Handle missing values
# Drop rows with missing values in specific columns
df_cleaned = df.dropna(subset=["FIRST_VACCINE_DATE","PERSONS_BOOSTER_ADD_DOSE_PER100","PERSONS_BOOSTER_ADD_DOSE","NUMBER_VACCINES_TYPES_USED","TOTAL_VACCINATIONS_PER100","TOTAL_VACCINATIONS"])
df_cleaned.head()
```

```python
# CHECKING FOR THE MISSING VALUES 
missing_values = df_cleaned.isnull().sum()
# DISPLAYING THE COUNTS OF MISSING VALUES FOR EACH VALUES
missing_values
```

```python
df_cleaned.head()
```

```python
df_cleaned["Fully_vaccinated"]= df_cleaned.loc[:,"PERSONS_VACCINATED_1PLUS_DOSE"]+df_cleaned.loc[:,"PERSONS_LAST_DOSE"]+df_cleaned.loc[:,"PERSONS_BOOSTER_ADD_DOSE"]
df_cleaned
```

```python
df_cleaned.head()
```

```python
#Ration of people fully vaciinated to total vaccination
df_cleaned["fully_vaccination_ratio"] = df_cleaned["Fully_vaccinated"]/df_cleaned["TOTAL_VACCINATIONS"]
df_cleaned.head()
```

```python
# Group the data by country and calculate the median vaccination coverage
median_coverage_by_country= df_cleaned.groupby("COUNTRY")["fully_vaccination_ratio"].median().reset_index()

## Display the median vaccination coverage by country
median_coverage_by_country
```

```python
# Filter the median coverage data for India
median_coverage_india = median_coverage_by_country[median_coverage_by_country["COUNTRY"]=="India"]
median_coverage_india
```

```python
# Group the data by country and calculate the median vaccination coverage
median_coverage_by_country_fully_vaccinated= df_cleaned.groupby("COUNTRY")["Fully_vaccinated"].median().reset_index()

## Display the median vaccination coverage by country
median_coverage_by_country_fully_vaccinated
```

```python
# Filter the median coverage data for India
median_coverage_india_fully_vaccination = median_coverage_by_country_fully_vaccinated[median_coverage_by_country["COUNTRY"]=="India"]
median_coverage_india_fully_vaccination
```

```python
 #Group the data by country and calculate the total vaccinations for each country
    
total_vaccinations_by_country = df_cleaned.groupby("ISO3")["TOTAL_VACCINATIONS"].max()

# Get the top 10 countries with the highest total vaccinations

top_10_countries = total_vaccinations_by_country.nlargest(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_10_countries.plot(kind="bar", color="skyblue")
plt.title("Top 10 Countries with the Highest Total Vaccinations")
plt.xlabel("Country")
plt.ylabel("Total Vaccinations")
plt.xticks(rotation=45)



plt.tight_layout()


# Display the plot
plt.show()
```

### What is the distribution of total vaccinations across top 10 countries? Create a bar plot to visualize the total vaccinations for each country.

```python
total_vaccination_by_country = df_cleaned.groupby('COUNTRY')["TOTAL_VACCINATIONS"].max()

plt.figure(figsize=(12,8))
total_vaccinations_by_country.nlargest(10).sort_values(ascending=False).plot( kind = 'bar', color="skyblue")
plt.title("Distribution of Total Vaccinations Across Different Countries")
plt.xlabel("Country")
plt.ylabel("Total Vaccinations")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### What is the correlation between fully vaccinated people and total vaccinations? Calculate the correlation coefficient between the

```python
# How has PERSONS_LAST_DOSE was last updated? Plot a line chart showing the daily vaccinations over time globally.
```

```python
df_cleaned["DATE_UPDATED"] = pd.to_datetime(df_cleaned["DATE_UPDATED"])
# Group the data by date and calculate the sum of daily vaccinations
daily_vaccinations_global = df_cleaned.groupby("DATE_UPDATED")["PERSONS_LAST_DOSE"].sum()
# Create a line chart
plt.figure(figsize=(12, 8))
plt.plot(daily_vaccinations_global.index, daily_vaccinations_global.values, marker='o', linestyle='-', color='b')
plt.title("DATE_UPDATED VS PERSON_LAST_DOSE(GLOBALLY)")
plt.xlabel("Date_UPDATED")
plt.ylabel("PERSONS_LAST_DOSE")
plt.xticks(rotation=45)
plt.tight_layout()
# Display the plot
plt.show()
```

```python
df_cleaned["DATE_UPDATED"] = pd.to_datetime(df_cleaned["DATE_UPDATED"])
# Gr
daily_vaccinations_global = df_cleaned.groupby("DATE_UPDATED")["Fully_vaccinated"].sum()
# Create a line chart
plt.figure(figsize=(12, 8))
plt.plot(daily_vaccinations_global.index, daily_vaccinations_global.values, marker='o', linestyle='-', color='b')
plt.title("DATE_UPDATED VS Fully_vaccinated(GLOBALLY)")
plt.xlabel("Date_UPDATED")
plt.ylabel("Fully_vaccinated")
plt.xticks(rotation=45)
plt.tight_layout()
# Display the plot
plt.show()
```

```python
# Filter out non-numerical columns
numerical_columns = df_cleaned.select_dtypes(include=["float64"]).columns
# Calculate the correlation matrix
corr_matrix = df_cleaned[numerical_columns].corr()
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Numerical Vaccination Metrics")
plt.show()
```
